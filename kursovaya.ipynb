–í –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω—ã m —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∏—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å b1,‚Äâb2,‚Äâ...,‚Äâbm (1‚Äâ‚â§‚Äâbi‚Äâ‚â§‚Äân).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ä–æ–≤–Ω–æ –æ–¥–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å ai, —Ç–∞–∫–∞—è —á—Ç–æ bi‚Äâ=‚Äâfai –¥–ª—è –≤—Å–µ—Ö i –æ—Ç 1 –¥–æ m, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ "Possible". –î–∞–ª–µ–µ –≤—ã–≤–µ–¥–∏—Ç–µ m —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª a1,‚Äâa2,‚Äâ...,‚Äâam.

–ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π ai, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ "Ambiguity".

```python
def solve_task_1():
    n, m = input().split(' ')
    n = int(n)
    m = int(m)
    list_f = input().split(' ')
    list_b = input().split(' ')
    for i in range(n):
        list_f[i] = int(list_f[i])
        list_b[i] = int(list_b[i])
    dict_b = {}
    dict_f = {}
    for i in range(n):
        if list_f[i] not in dict_f.keys():
            dict_f[list_f[i]] = 1
        else:
            dict_f[list_f[i]] += 1
    for i in range(m):
        if list_b[i] not in dict_b.keys():
            dict_b[list_b[i]] = 1
        else:
            dict_b[list_b[i]] += 1
    list_a = []
    is_one_solve = True
    keys = dict_b.keys() if len(dict_b) > len(dict_f) else dict_f.keys()
    for key in keys:
        if key not in dict_b.keys() or key not in dict_f.keys():
            print("Impossible")
            return
        else:
            if (dict_f[key] != dict_b[key]) or dict_f[key] > 1 or dict_b[key] > 1:
                is_one_solve = False
            else:
                list_a.append(key)
    if (is_one_solve):
        print("Possible")
        print(list_a)
    else:
        print("Ambiguity")
solve_task_1()
```

–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å s1,‚Äâ‚Äâs2,‚Äâ‚Äâ...,‚Äâ‚Äâsk –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–¥–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ b1,‚Äâ‚Äâb2,‚Äâ‚Äâ...,‚Äâ‚Äâbn, –µ—Å–ª–∏ –Ω–∞–π–¥–µ—Ç—Å—è —Ç–∞–∫–∞—è –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–Ω–¥–µ–∫—Å–æ–≤ i1,‚Äâi2,‚Äâ...,‚Äâik (1‚Äâ‚Äâ‚â§‚Äâ‚Äâi1‚Äâ‚Äâ<‚Äâ‚Äâi2‚Äâ‚Äâ<‚Äâ... ‚Äâ‚Äâ<‚Äâ‚Äâik‚Äâ‚Äâ‚â§‚Äâ‚Äân), —á—Ç–æ bij‚Äâ‚Äâ=‚Äâ‚Äâsj. –ò–Ω—ã–º–∏ —Å–ª–æ–≤–∞–º–∏, –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å s –º–æ–∂–µ—Ç –±—ã—Ç—å –ø–æ–ª—É—á–µ–Ω–∞ –∏–∑ b –ø—É—Ç–µ–º –≤—ã—á–µ—Ä–∫–∏–≤–∞–Ω–∏—è –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ4000). –í —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ n —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª b1,‚Äâb2,‚Äâ...,‚Äâbn (1‚Äâ‚â§‚Äâbi‚Äâ‚â§‚Äâ106).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –¥–ª–∏–Ω—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –ø–æ –¥–ª–∏–Ω–µ, –∏—Å–∫–æ–º–æ–π –ø–æ–¥–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.

–ü—Ä–∏–º–µ—Ä—ã

```python
def solve_task_2():
    n = int(input()) 
    array = input().split(' ')
    for i in range(n):
        array[i] = int(array[i])
    ans = 0
    dp = [-1]
    for diff in range(-100, 100, -1):
        cur = 0
        for i in range(1, n):
            if (array[i] - diff >= 1 and array[i] - diff <= 100):
                dp.append(sum[array[i] - diff] + 1)
        sum[array[i]] += dp[i]
        cur += dp[i]
    
        ans += cur
        ans -= n

    ans += n
    ans = 2**n - ans
    print(ans)
solve_task_2()
```

```python
def solve_task_3():
    str = input()
    m = int(input())
    for i in range(m):
        li, ri = map(int, input().split())
        q = str[li-1:ri-1]
        sub_s_len = 3
        if len(q) < sub_s_len:
            print("YES")
            continue
        i = 0
        while i < len(q):
            if i + sub_s_len <= len(q):
                s = q[i:i+sub_s_len]
                if "z" not in s or "y" not in s or "x" not in s:
                    print("NO")
                    break
            i += 1
        if i >= len(q):
            print('YES')
solve_task_3()
```

–ò–Ω–∞—á–µ, –≤—ã–≤–µ–¥–∏—Ç–µ ¬´Possible¬ª –∞ –∑–∞—Ç–µ–º ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëè1,ùëè2,‚Ä¶,ùëèùëõ (1‚â§ùëèùëñ‚â§ùëõ).

```python
def solve_task_4():
    n = int(input())
    string = input().split()
    a = {}
    for i in range(n):
        num = n - int(string[i])
        if num not in a.keys():
            a[num] = 1
        else:
            a[num] += 1
    hat = 1
    ans = ''
    count = 0
    for key in a.keys():
        for i in range(key):
            ans += str(hat) + " "
            count += 1
        hat += 1
        if count > n:
            print("Impossible")
            return
    print("Possible")
    print(ans)
solve_task_4()
```

```python
def solve_task_5():
    n=int(input())
    p=list(map(int,input().split()))
    for i in range(n):
        p[i]-=1
    color=[0]*n
    c=1
    s=[]
    for i in range(n):
        if color[i]==0:
            j,k=i,1
            color[i]=c
            while p[j]!=i:
                j=p[j]
                color[j]=c
                k+=1
            c+=1
            s.append(k)
    s.sort()
    if len(s)==1:
        x=s[0]**2
    else:
        x=(s[-1]+s[-2])**2
        for i in range(len(s)-2):
            x+=s[i]**2
    print(x)
solve_task_5()
```

```python
def solve_task_6():
    n = int(input())
    l = list(map(int,input().split()))
    r = list(map(int,input().split()))
    a = []
    for i in range(n):
        a.append(n - l[i] - r[i])
    for i in range(n):
        for j in range(i):
            l[i] -= (a[j] > a[i])

    for i in range(n):
        for j in range(i+1, n):
            r[i] -= (a[j] > a[i])
    possible = True
    for i in range(n):
        if l[i] != 0 or r[i] != 0:
            possible = False
            break
    if possible:
        print("YES")
        print(a)
    else:
        print("NO")
solve_task_6()
```

```python
import numpy as np
def solve_task_7():
  n, k = map(int, input().split())
  a= np.empty((n, k))
  deleter= np.full(n,0)
  solvability=True
  counter=0
  for i in range(0,n):
    line = list(map(int, input().split()))
    for j in range(0,k):
      a[i][j]=line[j]
      counter+=a[i][j]
      if(counter==k):
       deleter[i]=1
    counter=0
  for i in range(0,n):
    if deleter[i]==1:
      a = np.delete(a,(i), axis = 0)
      n-=1
  for i in range(0,k):
    counter=0
    for j in range(0,n):
      if a[j][i]==1:
        counter +=1
      if(counter>n/2):
        solvability=False
  if(solvability==False):
    print("No")
  else:
    print("Yes")
solve_task_7()
```

```python
def is_palindrom(s):
    reverse = s[::-1]
    for i in range(len(s)):
        if s[i] != reverse[i]:
            return False
    return True

def solve_task_8():
    string = input()
    n = len(string)
    count = 0
    for i in range(n-1):
        for j in range(i+1, n):
            s = string[i:j]
            if is_palindrom(s):
                for f in range(j, n):
                    for k in range(f, n):
                        s1 = string[j:k+1]
                        if is_palindrom(s1) and len(s1) > 0:
                            count += 1
    print(count)
solve_task_8()
```

```python
import copy
def check(p):
    bads = 0
    for i in range(len(p)):
        bads += p[i] != i
    return bads <= 2

def solve_task_9():
    n, m = map(int, input().split())
    matrix = []
    for i in range(n):
        matrix.append(list(map(int, input().split())))
    
    for i in range(m):
        for j in range(i):
            flag = True
            for k in range(n):
                p = copy.copy(matrix[k])
                if not check(p):
                    p[i], p[j] = p[j], p[i]
                    if not check(p):
                        flag = False
            if flag:
                print("YES")
                return
    print("No")
solve_task_9()
```

–ü—Ä–∏–º–µ—Ä—ã –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 2 1 2 ba ac –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 1 –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 3 1 3 1 aa ba ac –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 1 –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 2 5 5 bbb aaa –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ -1 –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ 2 3 3 aaa aa –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ -1 –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ –í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—å —Å—Ç—Ä–æ–∫—É 2 –∏–ª–∏ —Å—Ç—Ä–æ–∫—É 3. –ù–∞ —Ä–∞–∑–≤–æ—Ä–æ—Ç —Å—Ç—Ä–æ–∫–∏ 3 —Ç—Ä–∞—Ç–∏—Ç—Å—è –º–µ–Ω—å—à–µ —ç–Ω–µ—Ä–≥–∏–∏, –ø–æ—ç—Ç–æ–º—É –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º –æ—Ç–≤–µ—Ç–æ–º –±—É–¥–µ—Ç —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—å –µ—ë. –í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –æ–±–µ —Å—Ç—Ä–æ–∫–∏ –Ω–µ –∏–∑–º–µ–Ω—è—é—Ç—Å—è –ø–æ—Å–ª–µ —Ä–∞–∑–≤–æ—Ä–æ—Ç–∞ –∏ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –≤ –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ, –ø–æ—ç—Ç–æ–º—É –æ—Ç–≤–µ—Ç–æ–º —è–≤–ª—è–µ—Ç—Å—è ‚Äâ-‚Äâ1. –í —á–µ—Ç–≤—ë—Ä—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–±–µ —Å—Ç—Ä–æ–∫–∏ —Å–æ—Å—Ç–æ—è—Ç –∏–∑ –±—É–∫–≤ ¬´a¬ª, –Ω–æ –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ —Å—Ç—Ä–æ–∫–∞ ¬´aa¬ª –¥–æ–ª–∂–Ω–∞ —Ä–∞—Å–ø–æ–ª–∞–≥–∞—Ç—å—Å—è —Ä–∞–Ω—å—à–µ —Å—Ç—Ä–æ–∫–∏ ¬´aaa¬ª, –ø–æ—ç—Ç–æ–º—É –æ—Ç–≤–µ—Ç ‚Äâ-‚Äâ1.

```python
import numpy as np
def hardestTask():
  N=100001
  s=[[""]*2 for i in range(N)]
  dp=[[0]*2 for i in range(N)]
  n= int(input())
  c = np.full(n+2,0)
  for i in range(1,n+1):
    c[i]= input()
  for i in range(1,n+1):
    s[i][0]=input()
    s[i][1]=s[i][0]
    s[i][1]= s[i][1][::-1]
    for j in range(0,2):
      dp[i][j]=1e18
      for k in range(0,2):
        if(s[i][j]>=s[i-1][k]):
          dp[i][j]=min(dp[i][j],dp[i-1][k]+j*c[i])
  ans = min(dp[n][0],dp[n][1])
  if(ans<1e18):
   return(ans)
  return -1

print(hardestTask())
```
